```python
mylabels='Male','Female','Male and Female','Female and Male','Male and Male'

cmap=plt.get_cmap('Spectral')
colors=[cmap(i)for i in np.linspace(0,1,8)]
plt.figure(figsize=(16,7))

plt.title('Male and Female Distribution')
plt.pie(df['Gender'].value_counts(),labels=mylabels,autopct='%1.1f%%', shadow=True,colors=colors)
plt.show()
```

Male Percentage is High so we replace the Null value with Male
